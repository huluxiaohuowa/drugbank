```python
#step1
import pandas as pd
import networkx as nx
import numpy as np
from collections import Counter, defaultdict
import matplotlib.pylab as pylab
import matplotlib.pyplot as plt
```

```python
#step2
#setup the plotting
pylab.rcParams['figure.figsize'] = 128, 128


# read the main data source
indications = pd.read_excel("xlzong_drugs_indications.xlsx")
```

```python
#step3
From_To_ind = []
matrix = indications.ix[:,1:]
print type(matrix)
print matrix.index
print len(matrix)
```

```python
#step4
count = 0
From_To_ind = []
for i in range(matrix.shape[1]):
    
    temp = []
    temp1 = []
    temp2 = []
    for j in range(matrix.shape[0]):
        if (matrix[indications.columns.values[i+1]])[j] ==1 :
            #print targets.sparse[j]
            #print targets.columns.values[i+1]
            temp1.append(indications.sparse[j])
            temp2.append(indications.columns.values[i+1])
            temp = (zip(temp1,temp2))
            count +=1
    From_To_ind += temp
            
    #break        
print count

print From_To_ind 
print len(From_To_ind)
```

```python
##step 5

print From_To_ind[0] 
print From_To_ind[0][0] 
print From_To_ind[0][1]


edges_weights =   [[key[0], key[1], 1] for key in From_To_ind]
nodeA=[]
for _ in indications.columns.values:
    nodeA.append(_)
print len(nodeA)
print type(nodeA)
#del nodeA[0]
#print nodeA
nodeB=[]
for _ in indications.sparse:
    nodeB.append(_)
print len(nodeB)
print type(nodeB)  
#print nodeB

nodes = nodeA + nodeB 
print len(nodes)


nodeListA = nodeA
nodeListB = nodeB
edges =  From_To_ind


G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)
position = nx.spring_layout(G)

nx.draw_networkx_nodes(G,position, nodelist=nodeListA, node_size=550,node_color="b",with_labels=False)
nx.draw_networkx_nodes(G,position, nodelist=nodeListB, node_size=550,node_color="r",with_labels=False )
nx.draw_networkx_edge_labels(G,position,font_size=1,alpha=0.1,rotate=True,with_labels=False) 

nx.draw_networkx_edges(G,position,with_labels=False)
nx.draw_networkx_labels(G,position,with_labels=False)
plt.savefig("indications_graph3.png") 
plt.title("Graph of relationships in the indications database", fontsize=20) 
plt.clf()
plt.show()
```
